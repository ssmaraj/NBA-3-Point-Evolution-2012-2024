### Dropping Unnecessary Columns for 3-Point Analysis

Since this project focuses specifically on the evolution of the 3-point shot in the NBA, 
I removed columns unrelated to shooting performance, such as free throws, rebounds, steals, and personal fouls. 
Keeping only relevant variables (3-point attempts, makes, accuracy, total points, assists, turnovers, and team information) 
streamlines the dataset for cleaner analysis and visualization.
